```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("DBSCAN.csv")
```

```python
df.drop(columns = ["sr.no"], inplace = True)
```

```python
df.head()
```

```python
plt.scatter(df["A"], df["B"])
plt.xlabel("Feature A")
plt.ylabel("Feature B")
plt.title("Data Without any clustering")
plt.show()
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
```

## KMeans Clustering

```python
kmeans = KMeans(n_clusters = 3)
labels = kmeans.fit_predict(df)
```

```python
plt.scatter(df["A"], df["B"], c = labels)
plt.xlabel("Feature A")
plt.ylabel("Feature B")
plt.title("KMeans Clustering")
plt.show()
```

## Hierachical Clustering

```python
hc = AgglomerativeClustering(n_clusters = 3, linkage = "complete")
labels = hc.fit_predict(df)
```

```python
plt.style.use("ggplot")
plt.scatter(df["A"], df["B"], c = labels)
plt.xlabel("Feature A")
plt.ylabel("Feature B")
plt.title("Hierachical Clustering")
plt.show()
```

```python
db = DBSCAN(eps = 40, min_samples = 10)
labels = db.fit_predict(df)
```

```python
plt.style.use("ggplot")
plt.scatter(df["A"], df["B"], c = labels)
plt.xlabel("Feature A")
plt.ylabel("Feature B")
plt.title("Density-Based Spatial Clustering")
plt.show()
```
