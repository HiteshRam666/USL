```python
import pandas as pd 
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = [
    ['milk', 'onion', 'nutmeg', 'kidney beans', 'eggs', 'yogurt'],
    ['chips', 'bananas', 'nutmeg', 'eggs', 'curd', 'red bull'],
    ['pani puri', 'chips', 'maggie', 'eggs', 'peanuts', 'onion'],
    ['chicken', 'fish', 'bread', 'banana'],
    ['chicken', 'milk', 'yogurt', 'bread', 'chips'],
    ['rice', 'apple', 'chips']
]
```

# Step 1: Transform data using TransactionEncoder

```python
te = TransactionEncoder()
te_ary  = te.fit(data).transform(data)
df = pd.DataFrame(te_ary, columns = te.columns_)
```

# Step 2: Apply Apriori Algorithm

```python
frequent_patterns = apriori(df, min_support = 0.3, use_colnames = True)
```

# Step 3: Generate Association Rules

```python
rules = association_rules(frequent_patterns)
```
