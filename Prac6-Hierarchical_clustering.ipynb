```python
# Import necessary libraries
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt

# Import dataset and clustering algorithm
from sklearn.datasets import load_iris  # Load the Iris dataset
from sklearn.cluster import AgglomerativeClustering  # Perform hierarchical clustering

# Import functions for hierarchical clustering visualization
from scipy.cluster.hierarchy import dendrogram, linkage  # For creating dendrograms
```

```python
# Load the Iris dataset
iris = load_iris()  

# Extract the feature matrix (X) from the dataset
# 'iris.data' contains the numerical features of the flowers (sepal length, sepal width, petal length, petal width)
X = iris.data  
```

```python
# Perform hierarchical clustering using method
linked = linkage(X, method="complete")  
```

```python
# Create a figure with specified size
plt.figure(figsize=(10, 5))
# Plot the dendrogram using the hierarchical clustering result
dendrogram(linked)
# Add title and labels for better visualization
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data Points")  # X-axis represents the indices of data points
plt.ylabel("Distance")  # Y-axis represents the distance between clusters
# Display the dendrogram plot
plt.show()
```

```python
# Create an instance of AgglomerativeClustering with 3 clusters
hc = AgglomerativeClustering(n_clusters=3, metric = "euclidean")

# Fit the model to the dataset and predict cluster labels
labels = hc.fit_predict(X)
```

```python
# Set the figure size for better visibility
plt.figure(figsize=(8, 6))

# Create a scatter plot of the first two features (Sepal Length & Sepal Width)
# Color points based on cluster labels, use 'viridis' colormap, add black edges, and set point size
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k', s=400)

# Label the x-axis and y-axis
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

# Set the title of the plot
plt.title("Hierarchical Clustering on Iris Dataset")

# Display the plot
plt.show()
```

# Below code is Optional (Just for understanding) 

```python
for method in ["single", "complete", "average"]:
    linked = linkage(X, method)
    plt.figure(figsize = (10, 8))
    dendrogram(linked)
    plt.title(f"Dendrogram using {method} linkage")
    plt.xlabel("Data Points")
    plt.ylabel("Distance")
    plt.show()
```
